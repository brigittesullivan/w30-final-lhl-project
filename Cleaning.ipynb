```python
## Concat both DF into one data set
df_simple = pd.concat([df_1605_simple, df_0306_simple], axis = 0, ignore_index = True) # stack vertically
df_simple
```

```python
# perform cleaning set on combined data set
def cleaning_steps(df_name):
    # make lower case
    df_name['Commodity'] = df_name['Commodity'].str.lower()
    # keep only commodities that mention 'lobster'
    df_name = df_name[df_name['Commodity'].str.contains('lobster')].reset_index(drop=True)
    # drop extra columns
    df_name = df_name.drop(columns = ['Province', 'Country', 'State', 'Unit of measure'])
    # rename columns
    new_col_names = {'Value ($)': 'Value_$', 'Quantity': 'Quantity_kg'}
    df_name = df_name.rename(columns=new_col_names)
    # update period to datetime object
    df_name['Period'] = pd.to_datetime(df_name['Period'])
    return df_name

```

```python
df_simple = cleaning_steps(df_simple)
df_simple
```

```python
# drop commodity column
df_simple = df_simple.drop(columns=['Commodity'])
```

## Model with FB Prophet -Value($)

```python
def prophet_prep(df_name):
    df_name = df_name.rename(columns={ df_name.columns[0]: "ds" , df_name.columns[1]:'y'})
    return df_name
```

```python
df_value = prophet_prep(df_value)
```

```python
df_value
```

```python
m = Prophet()
```

```python
m.fit(df_value)
```

```python
future = m.make_future_dataframe(periods = 60, freq="MS")
```

```python
# future.tail(10)
```

```python
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
m.plot_components(forecast)
```
